##Eliminando valores nulos

```python
dataset = pd.read_csv('alquiler.csv', sep =';')
dataset.head()
```

```python
dataset.isnull()
```

```python
dataset.notnull()
```

```python
dataset.info()
```

```python
dataset[dataset['Valor'].isnull()]
```

```python
A = dataset.shape[0]
dataset.dropna(subset=['Valor'], inplace = True)
B = dataset.shape[0]
A - B
```

```python
dataset[dataset['Valor'].isnull()]
```

##Tratamientos Condicionados

```python
dataset.info()
```

```python
dataset[dataset['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (dataset['Mantenimiento'].isnull()) & (dataset['Tipo']=='Departamento')
```

```python
A = dataset.shape[0]
dataset = dataset[~seleccion]
B = dataset.shape[0]
A - B
```

```python
dataset[dataset['Mantenimiento'].isnull()].shape[0]
```

```python
dataset = dataset.fillna({'Mantenimiento': 0, 'Impuesto':0})
```

```python
dataset = dataset.fillna({'Mantenimiento': dataset['Mantenimiento'].mean(), 'Impuesto':dataset['Impuesto'].mean()})
```

```python
dataset.reset_index(drop = True, inplace= True)
```

```python
dataset.info()
```

#**5. Feature Engineering**

##Creando nuevas columnas

```python
dataset.info()
```

```python
dataset['Valor bruto'] = dataset.Valor + dataset.Mantenimiento + dataset.Impuesto
dataset.head()
```

```python
dataset['Valor m2'] = dataset.Valor / dataset.Area
dataset.head()
```

```python
dataset['Valor m2'] = dataset['Valor m2'].round(2)
dataset.head()
```

```python
dataset['Valor bruto m2'] = (dataset['Valor bruto'] / dataset.Area).round(2)
dataset.head()
```

```python
dataset.Tipo.value_counts()
```

```python
lista = dataset[dataset['Tipo'].str.contains('Casa')]['Tipo'].unique().tolist()
```

```python
dataset['Tipo Agrupado'] = dataset['Tipo'].apply(lambda x: 'Casa' if x in lista else 'Departamento')
dataset.head()
```

```python
dataset['Tipo Agrupado'].value_counts()
```

##Eliminando Columnas

```python
dataset.head()
```

```python
datos_aux = pd.DataFrame(dataset[['Valor bruto',	'Valor m2',	'Valor bruto m2',	'Tipo Agrupado']])
datos_aux.head()
```

```python
del datos_aux['Valor bruto']
datos_aux.head()
```

```python
#@title
import matplotlib.pyplot as plt
import numpy as np

# Crear datos para el boxplot con más outliers
np.random.seed(10)
# Datos normales
data = np.random.normal(0, 1, 100)

# Agregar outliers
outliers = np.random.uniform(low=-5, high=5, size=20)  # 20 valores atípicos
data = np.concatenate([data, outliers])

# Crear el boxplot
fig, ax = plt.subplots(figsize=(10, 5))

# Configuración del boxplot
ax.boxplot(data, vert=False, patch_artist=True,
           flierprops=dict(marker='o', color='black'),
           medianprops=dict(color='black', linewidth=2),
           boxprops=dict(facecolor='white', color='black', linewidth=2),
           whiskerprops=dict(color='black', linewidth=2),
           capprops=dict(color='black', linewidth=2))

# Añadir etiquetas para Q1, Q3, y la mediana
plt.text(-0.8, 1.1, 'Q1', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.text(0, 1.1, 'Mediana', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.text(0.8, 1.1, 'Q3', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')

# Añadir texto para la fórmula del IQR
plt.text(0, 0.88, 'IQR = Q3 - Q1', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')

# Añadir texto para los bigotes
plt.text(-2.5, 1.15, 'Q1 - 1.5 x IQR', horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')
plt.text(2.5, 1.15, 'Q3 + 1.5 x IQR', horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')

# Ajustar límites de x para que el boxplot esté más centrado
plt.xlim(-4, 4)

# Eliminar los ejes
ax.axis('off')
plt.text(0, 1.3, 'Boxplot', horizontalalignment='center', verticalalignment='center', fontsize=30, color='black')
plt.grid(False)
plt.show()
```

```python
valor = dataset['Valor']
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR
```

```python
seleccion = (valor>=limite_inferior) & (valor<=limite_superior)
dataset_new = dataset[seleccion]
```

##Visualizaciones Exploratorias

```python
area = plt.figure()
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
dataset_new.info()
```

```python
g1.scatter(dataset_new.Valor, dataset_new.Area)
g1.set_title('Valor x Area')

g2.hist(dataset_new.Valor)
g2.set_title('Histograma de Valor')

grupo = dataset_new[dataset_new['Tipo'].str.contains('Casa')].groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g3.bar(label, valores)
g3.set_title('Valor Promedio por Tipo de Casa')

corr = dataset_new.select_dtypes(include=['int64', 'float64']).corr()
g4.matshow(corr, cmap ='coolwarm')
g4.set_xticks(range(len(corr.columns)))
g4.set_yticks(range(len(corr.columns)))
g4.set_xticklabels(corr.columns, fontsize=8, rotation=90)
g4.set_yticklabels(corr.columns, fontsize=8)
g4.xaxis.set_ticks_position('bottom')
g4.xaxis.set_label_position('bottom')
```
